# 3. Handling Missing Data
Q5: Identify the columns in the DataFrame that contain missing values. Which function would you use to check for missing values in each column?

```python
df.isnull()
#df.isnull(): Returns a DataFrame of the same shape as df, where each entry is True if the original entry is NaN (missing), and False otherwise.
```

```python
print(df.isnull().sum())  # Checks for missing values in each column and sums them up
#Explanation:df.isnull().sum(): Sums up the True values in each column, showing the total number of missing values per column.
```

 Q6: How would you remove rows that contain any missing values from the DataFrame?

```python
df_cleaned = df.dropna()  # Removes rows with any missing values
print(df_cleaned)
#Explanation:df.dropna(): Removes all rows that contain at least one missing value. This is a quick way to clean the dataset, but be cautious as it can result in loss of data.
```

Q7: Instead of removing, how would you fill missing values in a column named AGE with the median value of that column?

```python
median_age = df['AGE'].median() 
median_age
```

```python
median_age = df['AGE'].median()  # Calculates the median of the 'AGE' column
df['AGE'].fillna(median_age, inplace=True)  # Fills missing values in 'AGE' with the median
#Explanation:df['AGE'].median(): Computes the median value of the AGE column. df['AGE'].fillna(median_age, inplace=True): Replaces all missing values in the Age column with the calculated median. The inplace=True argument modifies the DataFrame directly
```

# 5. Handling Duplicates
Q10: Write code to check for duplicate rows in the DataFrame. If duplicates exist, how would you remove them?

```python
duplicates = df.duplicated()  # Checks for duplicate rows
print(duplicates.sum())  # Prints the number of duplicate rows

df = df.drop_duplicates()  # Removes duplicate rows
#Explanation:
# df.duplicated(): Returns a boolean Series indicating whether each row is a duplicate of a previous row.
# df.drop_duplicates(): Removes all duplicate rows, keeping only the first occurrence of each duplicate.
```

Q11: After removing duplicates, how would you reset the DataFrame index?

```python
df.reset_index(drop=True, inplace=True)  # Resets the DataFrame index
#Explanation:
# df.reset_index(drop=True, inplace=True): Resets the index of the DataFrame after removing duplicates. The drop=True argument prevents the old index from being added as a column.
```

# 6. Updating and Modifying Rows
Q12: How would you update the value in the Discount column for all customers whose shipping mode is Regular Air by adding 10% to their current salary?

```python
df = pd.read_excel('Sample Superstore (1).xlsx') 
df.head()
```

```python
df.loc[df['Ship Mode'] == 'Regular Air', 'Discount'] *= 1.10  # Increases discount by 10% for customers based on their shipping mode

#Explanation:
# df.loc[df['Ship Mode'] == 'Regular Air', 'Salary']: Selects the rows where the Ship Mode is Regular Air and targets the Discount column.
# *= 1.10: Multiplies the selected discounts by 1.10, effectively increasing them by 10%.
```

```python
print(df.head())
```

Q13: Write a function that applies a 5% tax deduction to a column named Sales, and use this function to update the column.

```python
def tax_deduction(Sales):
    return Sales * 0.95  # Applies a 5% tax deduction

df['Sales'] = df['Sales'].apply(tax_deduction)  # Updates the 'Sales' column using the function
df['Sales']

#Explanation:

#def tax_deduction(salary): Defines a function that returns 95% of the input salary, simulating a 5% tax deduction.
#df['Salary'].apply(tax_deduction): Applies the tax_deduction function to each entry in the Salary column.
```

# 7. Filtering Data
Q14: Write code to filter the DataFrame to include only rows where the Shipping Cost column is greater than 2.

```python
df_filtered = df[df['Shipping Cost'] > 2]  # Filters the DataFrame to include only rows where 'Shipping Cost' > 2
df_filtered

#Explanation:

#df['Shipping Cost'] > 2: Creates a boolean mask where True indicates that Shipping Cost is greater than 2.
#df[df['Shipping Cost'] > 2]: Uses the boolean mask to filter the DataFrame, returning only rows that satisfy the condition.
```

Q15: How would you filter the DataFrame to include only rows where the column is either Customer Segment or Region?

```python
df_filtered = df[df.isin(['Customer Segment', 'Region'])]  # Filters the DataFrame to include only 'Customer Segment' and 'Region' departments
df_filtered
```
